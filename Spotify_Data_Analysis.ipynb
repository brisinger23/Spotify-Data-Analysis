```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_track= pd.read_csv('SpotifyFeatures.csv')
df_track.head()
```

```python
df_track.describe()
```

```python
df_track.info()
```

```python
df_track.isnull().sum()
```

```python
#checking top 10 most popular songs
df_most_popular_sorted= df_track.sort_values('popularity', ascending=False).head(10)
df_most_popular_sorted
```

```python
#checking 10 least popular songs
df_least_popular_sorted= df_track.sort_values('popularity', ascending=True).head(10)
df_least_popular_sorted
```

```python
#checking songs with more than 90% popularity

most_popular= df_track.query('popularity>90', inplace=False).sort_values('popularity', ascending=False)
most_popular[:10]
```

```python
df_track['duration']=df_track['duration_ms'].apply(lambda x: round(x/1000))
df_track.drop('duration_ms', inplace= True, axis=1)
```

```python
df_track.head()
```

```python
#visulization time
#correlation map

corr_df = df_track.drop(['key', 'mode'], axis=1).corr(method='pearson')

plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linecolor='black')
heatmap.set_title('Correlation Heatmap Between Variables')
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)

plt.show()
```

```python
#lets breakdown the dataset into only 4 percent
sample_df= df_track.sample(int(0.004*len(df_track)))
```

```python
#regression plot between loudness and energy

plt.figure(figsize=[10,6])
sns.regplot(data= sample_df, y='loudness', x='energy',color='c')
```

```python
#regression plot between popularity and acousticness of songs
plt.figure(figsize=[10,6])
sns.regplot(data= sample_df, y='popularity', x='acousticness',color='g')
```

```python
plt.figure(figsize=[15,10])
plt.title('Duration of the song in different genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(y='genre', x= 'duration', data=df_track)
```

```python
sns.set_style(style= 'darkgrid')
plt.figure(figsize=[15,10])
most_popular_genres= df_track.sort_values('popularity', ascending= False).head(10)
sns.barplot(y='genre',x='popularity', data= most_popular_genres).set(title='Top 10 genre by popularity')
```
